```python
import pandas as pd
import numpy as np
```

```python
# read clean data
rating_df = pd.read_csv('./data/clean/movie-lens-100k-ratings.csv').astype(str)
user_df = pd.read_csv('./data/clean/movie-lens-100k-users.csv').astype(str)
item_df = pd.read_csv('./data/clean/movie-lens-100k-items.csv').astype(str)

# get user, item dummies
item_dummies_df = pd.get_dummies(item_df.set_index('item_id'), dummy_na=False).astype(int)
user_dummies_df = pd.get_dummies(user_df.set_index('user_id')).astype(int)

# remove nan features
item_dummies_df = item_dummies_df.loc[:, ~item_dummies_df.columns.str.contains('nan')]
```

```python
# merge user, item dummies to rating data
data = rating_df.merge(user_dummies_df, left_on='user_id', right_index=True, how='left')\
    .merge(item_dummies_df, left_on='item_id', right_index=True, how='left')

# sort chronologically and set index
data = data.sort_values('timestamp', ascending=True).set_index(['user_id', 'item_id', 'timestamp'])
data.head()
```
